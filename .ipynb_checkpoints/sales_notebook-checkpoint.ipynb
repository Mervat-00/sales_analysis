```python
fig = plt.figure(figsize=(12, 8))
plt.plot(info['date'] , info['Returns'] , color='red' , label= 'Returns' , marker = '.');
plt.plot(info['date'] , info['Discounts'] , color='black' , label='Discounts' , marker = '.');
plt.xlabel('Date')
plt.ylabel('Loss')
plt.legend()
plt.tick_params(axis='x',labelsize=15,rotation=90)
plt.tight_layout()
```

```python
fig2 =plt.figure(figsize=(12,8))
plt.plot(info['date'] , info['Net Sales'] , color='green' , label='Net Sales',marker = '.');
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.tick_params(axis='x' , labelsize = 15 , rotation=90)
plt.tight_layout()
```

```python
fig3 = plt.figure(figsize =(12,8))
plt.plot(info['date'],np.array(info['Gross Sales']) , label ='Gross Sales' , color='orange' , linewidth=2 , marker = '.')
plt.xlabel('Date')
plt.ylabel('Gross Sales')
plt.legend()
plt.tick_params(axis ='x' , labelsize=15 ,rotation=90)
plt.tight_layout()
```
